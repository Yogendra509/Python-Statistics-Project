```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("SALARY_DATA.csv")
```

```python
print(df.head(),"\n")
print(df.tail(),"\n")
print(df.shape,"\n")
print(df)
```

#Basic Analysis

```python
#Average Salary
average_salary = df["Salary"].mean()

#Highest Salary
highest_salary = df["Salary"].max()

#Lowest Salary
lowest_salary = df["Salary"].min()

print(f"Average_Salary: ₹{average_salary:,.2f}"
     )
print(f"Highest_Salary: ₹{highest_salary:,.2f}")
print(f"Lowest_Salary: ₹{lowest_salary:,.2f}")
print(f"The average salary in the company is ₹{average_salary:,.2f}.")
print(f"The salary range is ₹{lowest_salary:.2f} to ₹{highest_salary:,.2f}.")
```

#Filtering

```python
#Employees earning above 70K
High_salary = df[df["Salary"] > 70000]

#Employees in IT Department
IT_Employee = df[df['Department'] == 'IT']

#HR Employees below 50K
HR_low_salary = df[(df["Department"] == "HR") & (df["Salary"] < 50000)]

#Percentage High Salary
percentage_High_salary = (len(High_salary) / len(df)) * 100

print(f"Employees eaning above 70K: {len(High_salary)}\n")
print('Employees earning above 70K\n',High_salary,'\n')
print(f"Empolyees in IT Department: {len(IT_Employee)}")
print('\nEmployees in IT Department:\n',IT_Employee,'\n')
print(f"HR Employees below 50K: {len(HR_low_salary)}")
print('\nHR Employees below 50K:\n',HR_low_salary,'\n')
print(f"{percentage_High_salary:.2f}% employees earn above 70K")
```
